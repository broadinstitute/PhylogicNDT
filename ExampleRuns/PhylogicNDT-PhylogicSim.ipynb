### List of output files provided

```python
ls
```

 - *\*_plot.png and _report.pdf* Summary plots and simulation parameters. Best way to visualize the results of the simulation
 - *\*_X.txt* MAF-like files to input into the Clustering module, where the X represents the sample number (0 indexed). 
 - *\*simulation_paramters.tsv* includes all the input parameters that were used to run this specific simulation itheration.
 - *\*.truth.txt* List of all the mutations and their true CN and multiplicity
